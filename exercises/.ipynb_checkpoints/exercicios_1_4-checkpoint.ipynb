# Exercicio 27 (ultimo)

```python
ap = pd.read_csv('https://www.ime.usp.br/~noproest/dados/areas.txt', sep=r'\s+')
ap.head()
```

```python
# a)
print(f"Número total de apartamentos: {len(ap)}")
print(f"Número de apartamentos no bloco A: {len(ap[ap['Bloco'] == 'A'])}")
print(f"Número de apartamentos no bloco B: {len(ap[ap['Bloco'] == 'B'])}")
print(f"Número de apartamentos por andar: {len(ap[(ap['Andar'] == 1) & (ap['Bloco'] == 'A')])}")
```

```python
# b)
colunas = ['Sala', 'Cozinha', 'Banheiro', 'Dorm']
fig, axes = plt.subplots(nrows=4,ncols=1,figsize=(5,10))

for i, coll in enumerate(colunas):
    # Criando tabelas de frequência
    minn = math.floor(ap[coll].min())
    maxx = math.ceil(ap[coll].max())

    bins = np.linspace(minn, maxx, 7)
    
    table = pd.cut(ap[coll], bins=bins, right=False).value_counts().reset_index().sort_values(by=coll)
    table['freq'] = round(table['count'] / table['count'].sum(), 2)
    
    print(f'{table}\n')

    # Criando plots
    axes[i].set_title(f'{coll}')
    axes[i].boxplot(ap[coll])

plt.tight_layout()
    
    
```

```python
# c)

colunas = ['Sala', 'Cozinha', 'Banheiro', 'Dorm']
fig, axes = plt.subplots(nrows=2,ncols=4,figsize=(10,5))

for j, bloco in enumerate(['A','B']):
    for i, coll in enumerate(colunas):
        # Criando tabelas de frequência
        minn = math.floor(ap[ap['Bloco'] == bloco][coll].min())
        maxx = math.ceil(ap[ap['Bloco'] == bloco][coll].max())
    
        bins = np.linspace(minn, maxx, 7)
        
        table = pd.cut(ap[ap['Bloco'] == bloco][coll], bins=bins, right=False).value_counts().reset_index().sort_values(by=coll)
        table['freq'] = round(table['count'] / table['count'].sum(), 2)
        
        print(f'{table}\n')
    
        # Criando plots
        axes[j][i].set_title(f'{coll} do bloco {bloco}')
        axes[j][i].boxplot(ap[ap['Bloco'] == bloco][coll])

        axes[j][i].set_yticks(np.linspace(ap[coll].min(), ap[coll].max(), 5))
        
plt.tight_layout()
```

```python
# d)
ap['Total'] = ap['Sala'] + ap['Cozinha'] + ap['Banheiro'] + ap['Dorm']
plt.figure(figsize=(5,3))
plt.boxplot(ap['Total'])
plt.show()
```

```python
# e)
totalA = ap[ap['Bloco'] == 'A']['Total'].sort_values()
totalB = ap[ap['Bloco'] == 'B']['Total'].sort_values()

linha = np.arange(40,65,5)
plt.plot(linha,linha,color='red',zorder=2)
plt.scatter(totalA, totalB,zorder=3)
plt.grid(linestyle='--', color='lightgray',zorder=0)

plt.title('Area Total entre Blocos')
plt.xlabel('Area Total Bloco A')
plt.ylabel('Area Total Bloco B')

plt.xticks(linha)
plt.yticks(linha)

plt.xlim(40, 60)
plt.ylim(40, 60)
```

```python
print(f"Média de área total do bloco A: {ap[ap['Bloco'] == 'A']['Total'].mean()}")
print(f"Média de área total do bloco B: {ap[ap['Bloco'] == 'B']['Total'].mean()}")
```

f) O bloco A apresenta uma diferença de área total bem considerável, podemos perceber visualmente através dos gráficos anteriores que a Sala e a Cozinha são os principais responsáveis pela diferença de área final entre os apartamentos de cada bloc, sendo a Sala com o nível mais extremo de diferença.
Além disso, alguns cômodos apresentam valores extremos no boxplot, tanto pra valores abaixo do espero quanto acima do esperado. Em destaque está: cozinha no bloco A, dorm. no bloco A, cozinha no bloco B e banheiro no bloco B.
